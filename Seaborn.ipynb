**#Seaborn**

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics with less code.

sns.jpg

```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#How to find the dataset present in the sns library

sns.get_dataset_names()
```

```python
#how to load dataset from the library

tip = sns.load_dataset("tips")
tip
```

```python
# Scatterplot

sns.scatterplot(x = "total_bill", y = "tip",data = tip)
plt.title("Tips plot")
plt.show()
```

```python
# Lineplot
sns.lineplot(x = "total_bill", y = "tip",data = tip)
plt.title("Tips plot")
plt.show()
```

```python
# Barplot

sns.barplot(x = "day", y = "total_bill",data = tip)
plt.title("day vs total bil")
plt.grid(alpha=0.3)
plt.show()
```

```python
# how to find the unique values in the dataset

tip.time.unique()
```

```python
# Histogram

sns.histplot(x = "total_bill",data = tip, kde = True)
plt.title("Tips plot")
plt.grid()
plt.show()
```

```python
#unique values in day column

tip.day.unique()
```

```python
# kde plot shows the distribution in the set

sns.kdeplot(x = "total_bill", data = tip)
plt.title("Tips plot")
plt.show()
```

```python
# Boxplot

sns.boxplot(x = "day",y = "total_bill" ,data = tip)
plt.title("Tips plot")
plt.show()
```

```python
# violinplot

sns.violinplot(x="day",y="total_bill",data=tip,color='Brown')
plt.title("day vs total bill")
plt.xlabel("day")
plt.ylabel("total bill")
plt.show()
```

```python
# pairplot

sns.pairplot(tip, hue= "sex")
```

```python
sns.pairplot(tip, hue= "day")
```

```python
# FacetGrid

g = sns.FacetGrid(tip, col= "time", row = "smoker")
g.map(sns.scatterplot,"total_bill","tip")
plt.show()
```

```python
sns.scatterplot(x = "total_bill",y = "tip", data=tip,hue = "time", style="smoker",size = "size")
plt.title("customize scatter plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(title ="Time",loc="right")
plt.show()
```
